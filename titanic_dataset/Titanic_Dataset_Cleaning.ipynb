```python
#import necessary libraries.
import pandas as pd
```

```python
#Read data from the csv file into the dataframe and set passengerId column s index.
df = pd.read_csv('Titanic-Dataset.csv', index_col='PassengerId')
df
```

```python
df.head(10)
```

```python
#df.shape
df.info()
```

```python
df2 = df.drop(columns=['Cabin', 'Embarked'])
df2.head(10)
```

```python
#Check number of unique names
df2['Name'].nunique()
```

```python
#Check whether Survived column contains only 0 or 1.
df2['Survived'].isin([0, 1]).value_counts()

#Checks whether Pclass column contains 1, 2, 3 only.
df2['Pclass'].isin([1, 2, 3]).value_counts()

#Checks whether Sex column contains only male or female.
df2['Sex'].isin(['male', 'female']).value_counts()

#Check whether SibSp column has null values.
df2['SibSp'].isnull().value_counts()

#Checks whether Parch column has null values.
df2['Parch'].isnull().value_counts()

#Checks whether the Fare column has null values.
df2['Fare'].isnull().value_counts()

#Checks whether the Ticket column has null values.
df2['Ticket'].isnull().value_counts()
```

```python
#Check number of null values in the Age column.
df2['Age'].isnull().value_counts()
df2['Age'].describe()
df2['Age'].mode()

#Fill Null Values of age with mean of all ages
df2['Age'].fillna(df2['Age'].mean(), inplace=True)

#Convert the age from float64 to int32
df2['Age'] =  df2['Age'].astype('int32')
```

```python
df2.head(10)
```

```python
df2['Ticket'] = df['Ticket']
df2.head(10)
```

```python
#Regex to remove the letters and values before the space from the ticket number
df2['Ticket'] = df2['Ticket'].str.replace(r'^\S+\s+', '', regex=True)
```

```python
df2.head(10)
df2.tail(10)
```

```python
df2[['Surname', 'First_Second_Name']] = df2['Name'].str.strip().str.split(',', n=1, expand=True)
df2.head(10)
```

```python
df2.columns
```

```python
df2[['Surname','First_Second_Name']].head(20)
```

```python
df2['First_Second_Name'] = df2['First_Second_Name'].str.replace(r'\s*\(.*\)', '', regex=True)
df2.head(20)
```

```python
df2.drop(columns='Name', inplace=True)
df2.to_csv('Cleaned_Titanic_Dataset.csv')
```
