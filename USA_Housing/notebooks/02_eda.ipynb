```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Housing_df = pd.read_csv('../data/USA_Housing.csv')
Housing_df.head(10)
```

## Univariate Analysis

```python
# Avg. Area Income vs Price

plt.figure(figsize=(10,8))
sns.scatterplot(x='Avg. Area Income', y='Price', data=Housing_df, alpha=0.5)
plt.title('Average Area Income vs Price', fontsize=14, weight='bold', pad=15)
plt.xlabel('Avg. Area Income', fontsize=12, labelpad=12)
plt.ylabel('Price', fontsize=12, labelpad=12)

plt.grid(axis='both', linestyle='--', alpha=0.7)
plt.tight_layout()

plt.savefig('../reports/figures/Avg. Income_Price.png')
```

```python
# Avg. Area Number of Rooms vs Price

plt.figure(figsize=(10,8))
sns.scatterplot(x='Avg. Area Number of Rooms', y='Price', data=Housing_df, alpha=0.5)
plt.title('Avg. Area Number of Rooms vs Price', fontsize=14, weight='bold', pad=15)
plt.xlabel('Avg. Area Number of Rooms', fontsize=12, labelpad=12)
plt.ylabel('Price', fontsize=12, labelpad=12)

plt.grid(axis='both', linestyle='--', alpha=0.7)
plt.tight_layout()

plt.savefig('../reports/figures/Avg. Rooms_Price.png')
```

## Correalational Analysis

```python
numerical_data = Housing_df.iloc[:,:-1]
numerical_data
```

```python
correlation = numerical_data.corr()
correlation
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(correlation, cmap='coolwarm', annot=True, vmin=-1, vmax=1)

plt.tight_layout()
plt.savefig('../reports/figures/Heatmap.png')
```
