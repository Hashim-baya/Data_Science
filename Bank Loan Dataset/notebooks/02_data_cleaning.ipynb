```python
import pandas as pd
pd.set_option('display.max_columns',300)
pd.set_option('display.max_rows',300)
```

```python
bank_df = pd.read_csv('../data/raw/Bank dataset 2.csv')
bank_df.head(10)
```

```python
print(list(bank_df.columns))
```

```python
column_list = ['customer_id', 'age', 'gender', 'marital_status', 'education_level', 'annual_income', 'loan_amount', 'credit_score', 'employment_years', 'employment_type', 'loan_purpose', 'home_ownership', 'debt_to_income_ratio', 'existing_loans', 'loan_term_months', 'interest_rate', 'loan_status', 'application_date',]
bank_data = bank_df[column_list].copy()
bank_data.head(10)
```

```python
bank_data.isnull().sum()
```

```python
bank_data.dropna(inplace=True)
bank_data.isnull().sum()
bank_data.reset_index(drop=True, inplace=True)
bank_data.head(10)
```

```python
bank_data.shape
```

```python
for column in bank_data.columns :
    print(f'....Describing {column}......:')
    print(bank_data[column].describe())
    print('\n')
```

```python
bank_data[['employment_years','age','existing_loans','loan_term_months','credit_score']] = bank_data[['employment_years','age','existing_loans','loan_term_months','credit_score']].astype('int64')
bank_data['application_date'] = pd.to_datetime(bank_data['application_date'])
bank_data.dtypes
bank_data.head(10)
```

```python
bank_data.to_csv('../data/interim/bank_data_cleaned.csv', index=False)
```
