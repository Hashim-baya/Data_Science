```python
import pandas as pd
pd.set_option('display.max_columns', 300)
pd.set_option('display.max_rows', 300)
```

```python
bank = pd.read_csv('../data/interim/bank_data_cleaned.csv')
bank.head(10)
```

```python
bank['Loan_to_Income_Ratio'] = round(bank['loan_amount'] / bank['annual_income'], 2)
bank.head()
```

```python
bank['monthly_income'] = round(bank['annual_income'] / 12, 2)

# Estimated amount already spent on existing debts monthly
bank['monthly_existing_debt_payment'] = round(bank['monthly_income'] * bank['debt_to_income_ratio'], 2)
bank.head()
```

```python
bank['employment_to_age_ratio'] = round(bank['employment_years'] / bank['age'], 2)
bank.head()
```

```python
# Flag high-obligation, subprime borrowers (Credit Score below Good tier)
bank['is_overextended_subprime'] = (
    (bank['existing_loans'] >= 3) & (bank['credit_score'] < 670)
).astype(int)
bank['is_overextended_subprime'].value_counts()
```

```python
bank.to_csv('../data/processed/Final_Bank_dataset.csv', index=False)
```
