```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns',300)
pd.set_option('display.max_rows',300)
```

```python
bank = pd.read_csv('../data/interim/bank_data_cleaned.csv')
bank.head(10)
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(x='loan_status',y='debt_to_income_ratio',data=bank,hue='employment_type',gap=0.1, palette='muted')

plt.title('Debt-to-Income (DTI) Ratio Distribution by Loan Status', fontsize=14, weight='bold', pad=15)
plt.xlabel('Loan Status', fontsize=12, labelpad=10)
plt.ylabel('Debt-to-Income (DTI) Ratio', fontsize=12, labelpad=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(title='Employment Type', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.savefig('../reports/figures/DTI_to_Loann_Status.png')
```

```python
plt.figure(figsize=(12,8))
sns.violinplot(x='loan_status', y='credit_score', data=bank, hue='employment_type', dodge=True, gap=0.1, palette='muted')

plt.title('Credit Score Distribution by Loan Status & Employment Type', fontsize=14, weight='bold', pad=15)
plt.xlabel('Loan Status', fontsize=12, labelpad=10)
plt.ylabel('Credit Score', fontsize=12, labelpad=10)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.legend(title='Employment Type', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.savefig('../reports/figures/Credit_Score_to_Loann_Status.png')
```

```python
plt.figure(figsize=(12,8))
sns.barplot(x='loan_status', y='existing_loans', data=bank,hue='gender', palette='muted', dodge=True, gap=0.1, errorbar=('ci', False))

plt.title('Existing Loans against Loan Status', fontsize=14, weight='bold', pad=15)
plt.xlabel('Loan Status', fontsize=12, labelpad=12)
plt.ylabel('Existing Loans', fontsize=12, labelpad=12)
plt.legend(title='Gender', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.savefig('../reports/figures/Existing_loans_to_loan_status.png')
```

```python
bank['month'] = bank['application_date'].map(lambda x: x.split('-')[1])
bank['month'] = bank['month'].astype('int32').sort_values()

plt.figure(figsize=(12,8))
sns.countplot(x='month', data=bank, color='orange', alpha=0.7)
plt.title('Loan Applications by months', fontsize=14, weight='bold', pad=15)
plt.xlabel('Months', fontsize=12, labelpad=12)
plt.ylabel('Number of applications', fontsize=12, labelpad=12)
plt.yticks(np.arange(0, 1200, 100))

plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.savefig('../reports/figures/Loan_Application_Months.png')
```
