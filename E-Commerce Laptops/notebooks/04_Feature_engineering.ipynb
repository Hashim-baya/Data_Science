```python
Laptops_df = pd.read_csv('../data/interim/Laptops_Cleaned.csv')
Laptops_df.head(10)
```

```python
# Laptops_df['Discounted Price'].value_counts()
# for value in Laptops_df['Discounted Price'] :
#     if value <= 150000.0:
#         print(f'{value} : <150K')
#         Laptops_df[['Price Band']] = '<150K'
#     elif 150000.0 <= value < 300000.0 : 
#         print(f'{value} : 150K - 300K')
#         Laptops_df['Price Band'] = '150K - 300K'
#     elif 300000.0 <= value <= 500000.0 : 
#         print(f'{value} : 300K - 500K')
#         Laptops_df['Price Band'] = '300K - 500K'
#     else : 
#         print(f'{value} : 500K+')
#         Laptops_df['Price Band'] = '500K+'

# Define the bin edges and the corresponding category labels
bins = [0, 150000, 300000, 500000, float('inf')]
labels = ['<150K', '150K - 300K', '300K - 500K', '500K+']

# Create the new column
Laptops_df['Price Band'] = pd.cut(Laptops_df['Actual Price'], bins=bins, labels=labels, right=False)
```

```python
Laptops_df.head(10)
```

```python
Laptops_df['Discount Amount'] = Laptops_df['Actual Price'] - Laptops_df['Discounted Price']
Laptops_df.head(10)
```

```python
Laptops_df.loc[Laptops_df['Saving(%)'] == 0]
```

```python
import numpy as np

Laptops_df['Processor Brand'] = 
    np.where(Laptops_df['Core'].str.contains('Ryzen', case=False, na=False), 'AMD',
    np.where(Laptops_df['Core'].str.contains('Core|Intel|Ultra', case=False, na=False), 'Intel',
    np.where(Laptops_df['Core'].str.contains('M1|M2|M3|M4', case=False, na=False), 'Apple', 'Other'))
)
```
