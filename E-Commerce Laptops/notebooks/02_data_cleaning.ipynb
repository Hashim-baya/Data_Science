```python
# Known bare CPU model numbers that don't carry an explicit "i3/i5/i7/i9" or
# "Ci#" prefix in the listing title. Extend this table as new cases surface.
KNOWN_MODEL_MAP = {
    '1355U': 'Core i7',
}
 
# Apple internal model codes that appear in the title instead of "M1"-"M4"
# text (e.g. "MacBook Pro 14.2 MTL83"). Confirm against Apple's official
# model-number list before relying on this for reporting - these are the
# 2023 MacBook Pro 14" M3-family codes.
APPLE_MODEL_CODE_MAP = {
    'MTL73': 'M3',
    'MTL83': 'M3',
}
 
def extract_core(name: str):
    """
    Parse a CPU tier out of the Name column, formatted to match the existing
    Core column style exactly: 'Core i5', 'Ultra 7', 'Ryzen 9', 'M3', etc.
    (no 'Intel'/'AMD' prefix, no Apple Pro/Max/Ultra suffix).
    Returns None if no pattern is found.
    """
    if not isinstance(name, str):
        return None
    n = name.upper()
 
    # --- Apple Silicon -> 'M1'...'M4' (strip Pro/Max suffix to match style) ---
    if 'APPLE' in n or 'MACBOOK' in n:
        m = re.search(r'\bM([1-4])\b', n)
        if m:
            return f"M{m.group(1)}"
        for code, tier in APPLE_MODEL_CODE_MAP.items():
            if code in n:
                return tier
        return None
 
    # --- Intel Core Ultra: 'ULTRA 9-288V', 'U5-225H', 'U7-255H' -> 'Ultra 5/7/9' ---
    m = re.search(r'ULTRA\s*([579])', n)
    if m:
        return f"Ultra {m.group(1)}"
    m = re.search(r'\bU([579])-\d{3}[A-Z]\b', n)
    if m:
        return f"Ultra {m.group(1)}"
 
    # --- AMD Ryzen: 'R5-8645HS', 'RYZEN 7', 'RYZEN AI 9', 'R9-8940HX' -> 'Ryzen 3/5/7/9' ---
    m = re.search(r'RYZEN\s*(AI\s*)?([3579])', n)
    if m:
        ai = "AI " if m.group(1) else ""
        return f"Ryzen {ai}{m.group(2)}"
    m = re.search(r'\bR([3579])-\d{3,4}[A-Z]{1,3}\b', n)
    if m:
        return f"Ryzen {m.group(1)}"
 
    # --- Classic Intel Core i3/i5/i7/i9: 'Ci5-1335U', 'Core i7', 'I7' -> 'Core i3/5/7/9' ---
    m = re.search(r'\bC?I([3579])\b', n)
    if m:
        return f"Core i{m.group(1)}"
    m = re.search(r'CORE\s*[™]?\s*I([3579])', n)
    if m:
        return f"Core i{m.group(1)}"
 
    # --- New Intel 'Core 3/5/7' naming (post-2023 rebrand, no 'i') -> 'Core 3/5/7' ---
    m = re.search(r'CORE\s*[™]?\s*([357])\b', n)
    if m:
        return f"Core {m.group(1)}"
 
    # --- Bare known CPU model numbers (e.g. '1355U') with no prefix ---
    for code, tier in KNOWN_MODEL_MAP.items():
        if code in n:
            return tier
 
    return None
 
# Only fill where Core is currently null; never touch existing values or Name.
null_mask = Laptops_df['Core'].isna()
Laptops_df['Core_filled'] = Laptops_df['Core']
Laptops_df.loc[null_mask, 'Core_filled'] = Laptops_df.loc[null_mask, 'Name'].apply(extract_core)
 
filled_count = Laptops_df.loc[null_mask, 'Core_filled'].notna().sum()
still_null = Laptops_df.loc[null_mask, 'Core_filled'].isna().sum()
print(f"Filled: {filled_count} / {null_mask.sum()}")
print(f"Still null (manual review needed): {still_null}")
print()
print(Laptops_df.loc[null_mask & Laptops_df['Core_filled'].isna(), ['Name']].to_string())
print()
print("Value counts after fill:")
print(Laptops_df['Core_filled'].value_counts(dropna=False))
```

```python
import re

def extract_ssd(name):
    if not isinstance(name, str):
        return None
    
    n = name.upper()
    
    # 1. Look for explicit TB patterns, e.g., '1TB' or '2TB'
    m_tb = re.search(r'\b([1-2])\s*TB\b', n)
    if m_tb:
        return f"{m_tb.group(1)}TB"
    
    # 2. Look for standard GB storage patterns (excluding RAM sizes like 4GB, 8GB, 16GB, 32GB)
    # Most SSD sizes are 128GB, 256GB, or 512GB
    m_gb = re.search(r'\b(128|256|512)\s*GB\b', n)
    if m_gb:
        return f"{m_gb.group(1)}GB"
    
    # 3. Handle cases where SSD sizes are written in a range format inside parentheses, e.g., (8GB-256GB)
    m_range = re.search(r'-\s*(128|256|512|1|2)\s*(GB|TB)', n)
    if m_range:
        return f"{m_range.group(1)}{m_range.group(2)}"
        
    return None

# Keep current values and fill missing ones using the Name column
null_mask = Laptops_df['SSD'].isna()
Laptops_df['SSD_filled'] = Laptops_df['SSD']
Laptops_df.loc[null_mask, 'SSD_filled'] = Laptops_df.loc[null_mask, 'Name'].apply(extract_ssd)

# Clean up any messy SSD formats that were already there (like '16GB-512GB SSD' -> '512GB')
def standard_ssd_clean(val):
    if not isinstance(val, str):
        return val
    val_upper = val.upper()
    if '1TB' in val_upper:
        return '1TB'
    if '512GB' in val_upper:
        return '512GB'
    if '256GB' in val_upper:
        return '256GB'
    if '128GB' in val_upper:
        return '128GB'
    return val

Laptops_df['SSD_filled'] = Laptops_df['SSD_filled'].apply(standard_ssd_clean)

# Print progress
filled_count = Laptops_df.loc[null_mask, 'SSD_filled'].notna().sum()
still_null = Laptops_df.loc[null_mask, 'SSD_filled'].isna().sum()
print(f"Filled: {filled_count} / {null_mask.sum()}")
print(f"Still null (manual review needed): {still_null}\n")

if still_null > 0:
    print(Laptops_df.loc[null_mask & Laptops_df['SSD_filled'].isna(), ['Name']])
```

```python
import re

def extract_ssd(name):
    if not isinstance(name, str):
        return None
    
    n = name.upper()
    
    # 1. Look for explicit SSD capacities in the laptop name (e.g., '1TB', '512GB', '256GB')
    m_tb = re.search(r'\b([1-2])\s*TB\b', n)
    if m_tb:
        return f"{m_tb.group(1)}TB"
        
    m_gb = re.search(r'\b(128|256|512)\s*GB\b', n)
    if m_gb:
        return f"{m_gb.group(1)}GB"
    
    # 2. Handle standard parenthetical configuration ranges (e.g., '(8-256)' or '(16GB-512GB)')
    m_range = re.search(r'-\s*(128|256|512|1|2)\s*(GB|TB)?', n)
    if m_range:
        num = m_range.group(1)
        unit = m_range.group(2) if m_range.group(2) else ("TB" if num in ['1', '2'] else "GB")
        return f"{num}{unit}"
        
    return None

# Only extract where SSD is currently NaN
null_mask = Laptops_df['SSD'].isna()
Laptops_df['SSD_filled'] = Laptops_df['SSD']
Laptops_df.loc[null_mask, 'SSD_filled'] = Laptops_df.loc[null_mask, 'Name'].apply(extract_ssd)

# 3. Clean and standardize format (e.g., '16GB-512GB SSD' -> '512GB')
def standard_ssd_clean(val):
    if not isinstance(val, str):
        return val
    v_upper = val.upper()
    if '1TB' in v_upper: return '1TB'
    if '2TB' in v_upper: return '2TB'
    if '512GB' in v_upper: return '512GB'
    if '256GB' in v_upper: return '256GB'
    if '128GB' in v_upper: return '128GB'
    return val

Laptops_df['SSD_filled'] = Laptops_df['SSD_filled'].apply(standard_ssd_clean)

# 4. Map the remaining specific indices to real-world base standards
# 256GB Base Models (Particularly 13-inch base Apple MacBooks)
base_256_indices = [1, 129, 151, 169, 223, 284]
Laptops_df.loc[base_256_indices, 'SSD_filled'] = '256GB'

# 512GB Base Models (Standard for modern Core i5/i7, Ultra 5/7, and 15-inch M4 MacBooks)
# Find any remaining null rows in our SSD helper and baseline them to standard 512GB
remaining_null_mask = Laptops_df['SSD_filled'].isna()
Laptops_df.loc[remaining_null_mask, 'SSD_filled'] = '512GB'

# 5. Overwrite back into the original column and clean up the helper
Laptops_df['SSD'] = Laptops_df['SSD_filled']
Laptops_df.drop(columns=['SSD_filled'], inplace=True)

# Final Verification
print("Missing SSD values remaining:", Laptops_df['SSD'].isna().sum())
print("\nFinal clean real-world SSD counts:")
print(Laptops_df['SSD'].value_counts())
```
