```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', 300)
pd.set_option('display.max_rows', 300)
```

```python
Laptops = pd.read_csv('../data/processed/Final_Laptops_df.csv')
Laptops.head(10)
```

 ## Which brands offer the deepest average discounts (`Discount_Pct`)? Which offer the least?

```python
Savings_Average = Laptops.groupby('Brand')['Saving(%)'].mean()
Savings_Average
```

```python
plt.figure(figsize=(12,8))
sns.barplot(data=Savings_Average.reset_index(), x='Brand', y='Saving(%)')
plt.title('Average Discounts(%) of Brands', fontsize=14, pad=15, weight='bold')
plt.ylabel('Average Saving (%)', fontsize=12, labelpad=12)
plt.xlabel('Brand', fontsize=12, labelpad=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.savefig('../reports/Figures/Average_Discounts.png')
```

## What is the brand market share within this catalog (count of listings per brand)?

```python
plt.figure(figsize=(12,8))
sns.countplot(data=Laptops, x='Brand', order=sorted(Laptops['Brand'].unique()), color='orange', alpha=0.7)
plt.title('Listing Counts per Brand', fontsize=14, weight='bold', pad=15)
plt.xlabel('Brand', fontsize=12, labelpad=12)
plt.ylabel('No. of Listings', fontsize=12, labelpad=12)

plt.yticks(np.arange(0,120,10))
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.savefig('../reports/Figures/Listings_per_Brand.png')
```

## How does average/median price differ by brand?

```python
Price = Laptops.groupby('Brand')['Actual Price'].agg(['mean','median'])
Price
```

```python
plt.figure(figsize=(12,8))
sns.barplot(data=Price.reset_index(), x='Brand', y='mean', color='green', alpha=0.7)
plt.title('Brands against Mean Price', fontsize=14, weight='bold', pad=15)
plt.xlabel('Brand', fontsize=12, labelpad=12)
plt.ylabel('Mean', fontsize=12, labelpad=12)

plt.yticks(np.arange(0,650000,50000))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.savefig('../reports/Figures/Brands_Mean_Price.png')
```

## Which is the most used core tier and what type of SSD.

```python
plt.figure(figsize=(12,8))
sns.countplot(data=Laptops, y='Core', hue='SSD', palette='muted', dodge=True, gap=0.1)
plt.title('Most Used Core Tiers with their SSD', fontsize=14, weight='bold', pad=15)
plt.xlabel('Count', fontsize=12, labelpad=12)
plt.ylabel('Core Tier', fontsize=12, labelpad=12)
plt.xticks(rotation=90)
plt.legend(title='SSD', bbox_to_anchor=(1.05,1), loc='upper left')

plt.grid(axis='x', linestyle='--', alpha=0.8)
plt.tight_layout()
plt.savefig('../reports/Figures/Core_Tier_vs_SSD.png')
```
