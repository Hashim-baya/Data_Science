```python
HealthCare_df = pd.read_csv('../data/raw/healthcare_biotech_patient_dataset.csv')
HealthCare_df.head(15)
```

## Removing Duplicates

```python
HealthCare = HealthCare_df.drop_duplicates().copy()
```

```python
HealthCare.duplicated().sum()
```

## Handling Null Values

```python
HealthCare.isnull().sum()
```

```python
HealthCare[['gender', 'ethnicity']] = HealthCare[['gender', 'ethnicity']].fillna('Other')
```

```python
HealthCare['insurance_type'] = HealthCare['insurance_type'].fillna('Uninsured')
```

```python
HealthCare['income_bracket'] = HealthCare['income_bracket'].fillna('Undisclosed')
```

```python
HealthCare[['height_cm', 'weight_kg']] = HealthCare[['height_cm', 'weight_kg']].fillna(HealthCare[['height_cm', 'weight_kg']].mean())
```

```python
BMI = HealthCare['weight_kg'] / pow((HealthCare['height_cm'] / 100), 2)
HealthCare['bmi'] = HealthCare['bmi'].fillna(BMI)
```

```python
HealthCare['systolic_bp'] = HealthCare['systolic_bp'].fillna(HealthCare.groupby('age')['systolic_bp'].transform('median'))
```

```python
HealthCare['diastolic_bp'] = HealthCare['diastolic_bp'].fillna(HealthCare.groupby('age')['diastolic_bp'].transform('median'))
```

```python
# Drop rows only if ALL three primary lipid metrics are missing
HealthCare.dropna(subset=['cholesterol_total', 'ldl', 'hdl'], how='all', inplace=True)
```

```python
# Fill missing values using the median of matching age cohorts
for col in ['cholesterol_total', 'ldl', 'hdl']:
    HealthCare[col] = HealthCare[col].fillna(HealthCare.groupby('age')[col].transform('median'))
```

```python
# Drop rows ONLY if fasting glucose is missing
HealthCare.dropna(subset=['glucose_fasting'], inplace=True)
```

```python
HealthCare['hba1c'] = (HealthCare['glucose_fasting'] + 46.7) / 28.7
```

```python
HealthCare['smoking_status'] = HealthCare['smoking_status'].fillna('Unknown')
```

```python
HealthCare['alcohol_consumption'] = HealthCare['alcohol_consumption'].fillna('Unknown')
```

```python
HealthCare['physical_activity_level'] = HealthCare['physical_activity_level'].fillna('Unknown')
```

```python
# Drop rows ONLY if both WBC and RBC are missing entirely
HealthCare.dropna(subset=['white_blood_cell_count', 'red_blood_cell_count'], how='all', inplace=True)
```

```python
# RBC must use gender because male/female reference ranges differ significantly
HealthCare['red_blood_cell_count'] = HealthCare['red_blood_cell_count'].fillna(
    HealthCare.groupby(['gender', 'age'])['red_blood_cell_count'].transform('median')
)
```

```python
# WBC can be filled using age-group medians
HealthCare['white_blood_cell_count'] = HealthCare['white_blood_cell_count'].fillna(
    HealthCare.groupby('age')['white_blood_cell_count'].transform('median')
)
```

```python
HealthCare.dropna(subset=['red_blood_cell_count'], inplace=True)
```

```python
# Fill missing platelets using the median of matching age cohorts
HealthCare['platelet_count'] = HealthCare['platelet_count'].fillna(
    HealthCare.groupby('age')['platelet_count'].transform('median')
)
```

```python
# Create temporary WBC tiers to calculate realistic CRP levels
HealthCare['wbc_tier'] = pd.qcut(HealthCare['white_blood_cell_count'], q=4, labels=['Low', 'Normal', 'High', 'Very High'])

# Fill missing CRP using the median of patients in the same immune tier
HealthCare['crp_level'] = HealthCare['crp_level'].fillna(
    HealthCare.groupby('wbc_tier', observed=True)['crp_level'].transform('median')
)

# Drop the temporary tier column
HealthCare.drop(columns=['wbc_tier'], inplace=True)
```

```python
# Create a mask that excludes the 'Other' group from the calculation
known_ethnicities = HealthCare[HealthCare['ethnicity'] != 'Other']

# Calculate the medians using only known backgrounds
ethnic_medians = known_ethnicities.groupby(['ethnicity', 'age'])['vitamin_d_level'].median()

# Create the multi-index template for alignment
multi_index = ['ethnicity', 'age']

# Set the index temporarily to match ethnic_medians, fill the nulls, and reset it back
HealthCare['vitamin_d_level'] = (
    HealthCare.set_index(multi_index)['vitamin_d_level']
    .fillna(ethnic_medians)
    .values
)

# Final safety fallback using just age for anyone left over
HealthCare['vitamin_d_level'] = HealthCare['vitamin_d_level'].fillna(
    HealthCare.groupby('age')['vitamin_d_level'].transform('median')
)
```

```python
# Fix Sparsely Ordered Specialty Markers with an explicit fallback chain
for col in ['biomarker_a', 'biomarker_b']:
    if col in HealthCare.columns:
        # Step 1: Try filling with specific gender + age medians
        gender_age_median = HealthCare.groupby(['gender', 'age'])[col].transform('median').round()
        HealthCare[col] = HealthCare[col].fillna(gender_age_median)
        
        # Step 2: Widened Fallback - Fill any remaining empty slices using just age medians
        age_median = HealthCare.groupby('age')[col].transform('median').round()
        HealthCare[col] = HealthCare[col].fillna(age_median)
        
        # Step 3: Global Safety Net - Fill remaining edge cases with the global rounded median
        global_median = HealthCare[col].median()
        # If the whole column is empty (edge case), default to 0 to prevent crashes
        if pd.isna(global_median):
            global_median = 0
        HealthCare[col] = HealthCare[col].fillna(round(global_median))
```

```python
HealthCare['medication_count'] = HealthCare['medication_count'].fillna(0.0)
```

## Analysis of each column individually

```python
# 1. Calculate the standard Quartiles
Q1 = HealthCare['age'].quantile(0.25)
Q3 = HealthCare['age'].quantile(0.75)
IQR = Q3 - Q1

# 2. Define the standard statistical fences
lower_fence = Q1 - 1.5 * IQR
upper_fence = Q3 + 1.5 * IQR

print(f"Statistical bounds for Age: {lower_fence} to {upper_fence}")

# 3. Separate your clean rows from your outlier rows
age_outliers = HealthCare[(HealthCare['age'] < lower_fence) | (HealthCare['age'] > upper_fence)]
print(f"Number of rows flagged as age outliers: {len(age_outliers)}")
```

```python
# Safe Alternative: Cap extreme ages at a maximum logical threshold of 110 rather than deleting patients
HealthCare['age'] = HealthCare['age'].clip(lower=0, upper=110)
```

```python
# Step 1: Drop impossible administrative data entry typos (negative days)
HealthCare = HealthCare[HealthCare['length_of_stay_days'] >= 0]

# Step 2: NOW run the log transformation safely
log_los = np.log1p(HealthCare['length_of_stay_days'])

# Step 3: Calculate IQR on the log scale
q1 = log_los.quantile(0.25)
q3 = log_los.quantile(0.75)
iqr = q3 - q1

# Step 4: Calculate the upper fence and convert it back to raw days
log_upper_fence = q3 + (1.5 * iqr)
raw_upper_fence = np.expm1(log_upper_fence)

print(f"Any stay longer than {raw_upper_fence:.1f} days is a statistical outlier.")

# Step 5: Filter out rows above this logical threshold
HealthCare = HealthCare[HealthCare['length_of_stay_days'] <= raw_upper_fence]
```
