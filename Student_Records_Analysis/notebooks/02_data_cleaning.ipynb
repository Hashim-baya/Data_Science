```python
import pandas as pd
```

# Cleaning Students Dataset.

```python
students_df = pd.read_csv('../data/raw/students.csv', index_col='student_id')
students_df.head(10)
```

```python
students_df.isnull().sum()
```

```python
students = students_df.dropna()
```

```python
students.head(10)
```

```python
students.tail(10)
```

```python
students.isnull().sum()
```

```python
students[students.duplicated()]
```

```python
students.drop_duplicates(inplace=True)
```

```python
students.duplicated().sum()
```

```python
students.head(10)
```

```python
students[['first_name','last_name','email']] = students[['first_name','last_name','email']].map(lambda column: column.strip())
```

```python
students[['first_name','last_name','email']].head(10)
```

```python
students['gender'] = students['gender'].str.replace(r'^(Female|female|f)$', 'F', regex=True)
students['gender'] = students['gender'].str.replace(r'^(male|Male|m)$', 'M', regex=True)
```

```python
students.head(10)
```

```python
students.tail(15)
```

```python
students.to_csv('../data/interim/students_cleaned.csv')
```

# Cleaning Enrollments Dataset.

```python
enrollments_df = pd.read_csv('../data/raw/enrollments.csv', index_col='enrollment_id')
enrollments_df.head(10)
```

```python
enrollments_df.isnull().sum()
```

```python
enrollments_df.duplicated().sum()
enrollments_df[enrollments_df.duplicated()]
```

```python
enrollment = enrollments_df.drop_duplicates()
enrollment.duplicated().sum()
```

```python
enrollment['student_id'].nunique()
```

```python
enrollment['course_id'].nunique()
```

```python
invalid_rows = enrollment[(enrollment['score'] < 0) | (enrollment['score'] > 100)].index
enrollment = enrollment.drop(invalid_rows)
```

```python
enrollment.head(10)
```

```python
enrollment.describe()
```

```python
enrollment.info()
```

```python
enrollment.tail(10)
```

```python
enrollment.to_csv('../data/interim/enrollments_cleaned.csv')
```
